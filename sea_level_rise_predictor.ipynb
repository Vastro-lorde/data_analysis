```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
def draw_plot():
    # Read data from file
    df = pd.read_csv('./data/epa-sea-level.csv')

    x = df['Year']
    y = df['CSIRO Adjusted Sea Level']

    # Create scatter plot
    fig, ax = plt.subplots(figsize=(12,12))
    ax = plt.scatter(x, y)
    # Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit. 
    # Plot the line of best fit over the top of the scatter plot. 
    # Make the line go through the year 2050 to predict the sea level rise in 2050.
    line_regres = linregress(x, y)
    x_forcast_years = pd.Series(([i for i in range(x.min(), 2051)]))
    y_forcast_years = line_regres.slope * x_forcast_years + line_regres.intercept

    # Create first line of best fit
    plt.plot(x_forcast_years, y_forcast_years, 'r')


    # Create second line of best fit
    df_2000 = df.loc[df["Year"] >= 2000]
    x2 = df_2000["Year"]
    y2 = df_2000["CSIRO Adjusted Sea Level"]
    line_regres2000 = linregress(x2, y2)
    x_forcast_years2 = pd.Series(([i for i in range(x2.min(), 2051)]))
    y_forcast_years2 = line_regres2000.slope * x_forcast_years2 + line_regres2000.intercept
    plt.plot(x_forcast_years2, y_forcast_years2, 'b')

    # Add labels and title
    plt.title("Rise in global sea level")
    plt.xlabel("Year")
    plt.ylabel("Sea Level (inches)")

    
    

draw_plot()
```
